```python
import numpy as np
import matplotlib.pyplot as plt
```

<h3><em>Temperature and Pressure values at a resolution of 0.1</em></h3>

```python
temperature = np.arange(-30, 30+0.1, 0.1) #temperature values from -30 to 30 at a resolution of 0.1
pressure = np.arange(200, 1000+0.1,1) #pressure values from 200 to 1000 at a resolution of 1
temperature_K = temperature + 273.15 #converting temperature to kelvin
```

<h3><em>potential temperature function</em></h3>

```python
def potential_temperature(po=1000, rd=287.05, cpd=1005.2):
    t = temperature_K.reshape(1, 601) # reshaped the temperature values to 2D
    p = pressure.reshape(801,1) #reshaped the pressure values to 2D
    theta = t*(po/p)**(rd/cpd) # calculated for potential temperature in 2D
    return theta
```

```python
potential_temperature = potential_temperature()
potential_temperature
```

<h2><em>Plot</em></h2>

```python
fig, ax = plt.subplots(figsize=(8, 9))
vals = [1000, 900, 800, 700, 600, 500, 400, 300, 200]
levels = np.arange(260, 480, 10)

cax = ax.contour(temperature, pressure, potential_temperature, levels=levels, colors='green')
ax.set_xlabel('Temperature', fontsize=14)
ax.set_ylabel('Pressure', fontsize=14)
ax.clabel(cax, levels[::2], fmt='%d', fontsize=12) #Katie helped me with this part.
ax.set_yscale('log') 
ax.set_ylim(1000, 200)
ax.set_yticks(ticks=vals, labels=vals)
ax.grid(visible=True, linestyle=':', linewidth=1)
ax.set_title("ATSC5010 oteng, Lab 3: Emagram", fontsize=16, fontweight='bold', pad=13)

# QUESTION TWO
# =====================================================
ax.plot(0, 700, color='red', marker='.', markersize=11)  #pressure at 700mb
ax.plot(0, 900, color='blue', marker='.', markersize=11) # presssure at 900mb
ax.vlines(x=0, ymin=900, ymax=700, color='black') #vertical line joining Q1 and Q2
ax.text(1, 710, 'Q1', style='italic', fontsize=13) #added text(Q1) to indicated point
ax.text(1, 910, 'Q2', style='italic', fontsize=13) #added text(Q2) to indicated point


# QUESTION SIX
# =====================================================
ax.plot(10, 500, color='red', marker='.', markersize=11) #temperature at 10 degree celcius
ax.plot(-20, 500, color='blue', marker='.', markersize=11) #temperature at -20 degree celcius
ax.hlines(y=500, xmin=-20, xmax=10, color='black') # horizontal line joining Q5 and Q6
ax.text(8.5, 487, 'Q5', style='italic', fontsize=13) #added text(Q5) to indicated point
ax.text(-21, 487, 'Q6', style='italic', fontsize=13) #added text(Q6) to indicated point


# QUESTION TEN
# =====================================================
ax.plot(0, 800, color='red', marker='.', markersize=11) #pressure at 800mb
ax.plot(-10, 700, color='blue', marker='.', markersize=11) # pressure at 700mb
ax.text(-3.5, 825, 'Q9', style='italic', fontsize=13) #added text(Q9) to indicated point
ax.text(-15, 720, 'Q10', style='italic', fontsize=13); #added text(Q10) to indicated point
ax.plot([0, -10], [800, 700], color='black') # line joining Q9 and Q10

#Plot and save figure to preferred location
fig.savefig('lab3.jpg', dpi=100)
```
